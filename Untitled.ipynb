```python
import cv2
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans 
import matplotlib.pyplot as plt
from PIL import Image
from sklearn_extra.cluster import KMedoids
```

```python
image = plt.imread('large.jpg') 
plt.figure(figsize=(4,6))
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
image=Image.open('large.jpg')
image=image.resize((100,100))
image=np.array(image)
image=np.array(image)
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
im=image/255
pic=im.reshape(im.shape[0]*im.shape[1],im.shape[2])
```

```python
fig, ax = plt.subplots(2, 3, figsize=(16, 8))
count=1
for i in range(2):
    for j in range(3):

        kmeans = KMeans(n_clusters=count+1, random_state=0).fit(pic)
        pic_print = kmeans.cluster_centers_[kmeans.labels_]
        clustered_pic=pic_print.reshape(im.shape[0],im.shape[1],im.shape[2])
        count+=1
        ax[i][j].set_title('Cluster '+str(count))
        ax[i][j].imshow(clustered_pic)
        ax[i][j].set_axis_off()
plt.show()
```

```python
fig, ax = plt.subplots(2, 3, figsize=(16, 8))
count=1
for i in range(2):
    for j in range(3):

        kmedoids = KMedoids(n_clusters=count+1, random_state=0).fit(pic)
        pic_print = kmedoids.cluster_centers_[kmedoids.labels_]
        clustered_pic=pic_print.reshape(im.shape[0],im.shape[1],im.shape[2])
        count+=1
        ax[i][j].set_title('Cluster '+str(count))
        ax[i][j].imshow(clustered_pic)
        ax[i][j].set_axis_off()
plt.show()
```
